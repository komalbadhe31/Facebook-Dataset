#Data preparation

```python
adata.columns
```

```python
adata.shape
```

```python
adata.isnull().sum()
```

```python
ozone_mean=adata['Ozone'].mean()
adata['Ozone']= adata['Ozone'].fillna(ozone_mean) 


solar_mean = adata['Solar.R'].mean()
adata['Solar.R'] = adata['Solar.R'].fillna(solar_mean)
```

```python
adata.isnull().sum()
```

#Scatter plot

```python

plt.xlabel('Wind')
plt.ylabel('Temprature')
plt.title('This is scatter plot')
plt.scatter(adata['Wind'],adata['Temp'],c='red')
```

#swarm plot

```python
ax = sns.swarmplot(adata.Ozone, adata.Wind, data=adata, zorder=0)

sns.boxplot(adata.Ozone, adata.Wind, data=adata,
            showcaps=False, boxprops={'facecolor': 'None'},
            showfliers=False, whiskerprops={'linewidth': 0}, ax=ax)

plt.show()
```
